# Import Libraries

```python
import pandas as pd
import numpy as np
```

# Import dataset

```python
data = pd.read_csv('AB_NYC_2019.csv')
```

```python
data
```

# Show info

```python
data.info()
```

# Show null value count

```python
data.isnull().sum()
```

```python
data = data.drop_duplicates()
```

# Drop Unnecessary Columns

```python
data = data.drop(columns = ['name', 'neighbourhood', 'calculated_host_listings_count'])
```

```python
data.head()
```

# Converting Data Types and Removing Null Values

```python
data['host_name'] = data['host_name'].astype(str)
data['host_name'] = data['host_name'].str.replace('nan', ' ')
```

# Removing all Non-Alphabetical Characters

```python
data['host_name'] = data['host_name'].str.replace('[^a-zA-Z]', '', regex = True)
```

# Inserting Space Between Any 2 Names

```python
data['host_name'] = data['host_name'].str.replace(r'([a-z])([A-Z])', r'\1 \2', regex = True)
```

```python
data.head()
```

# Obtaining Priced AirBNB Data

```python
data_priced = data[data['price'] != 0]
```

```python
data_priced.head()
```

# Data where minimum nights is a year or less

```python
data_year = data[data['minimum_nights'] <= 365]
```

```python
data_year.head()
```

# Data where availability is at least a day

```python
data_available = data[data['availability_365'] != 0]
```

```python
data_available.head()
```

# Cleaning last review column

### Converting data type and removing null values

```python
data['last_review'] = data['last_review'].astype(str)
data['last_review'] = data['last_review'].str.replace('nan', ' ')
```

```python
data.head()
```

# Cleaning Reviews per Month Column

### Converting data type and removing null values

```python
data['reviews_per_month'] = data['reviews_per_month'].astype(str)
data['reviews_per_month'] = data['reviews_per_month'].str.replace('nan', ' ')
```
